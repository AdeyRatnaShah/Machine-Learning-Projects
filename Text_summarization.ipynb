```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df  = pd.read_csv('text_summarization_dataset.csv')
```

```python
df
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

# Tokenize the documents and summaries
tokenizer = Tokenizer(num_words=50000)  # Limit vocab size
tokenizer.fit_on_texts(df['document'])
document_sequences = tokenizer.texts_to_sequences(df['document'])
summary_sequences = tokenizer.texts_to_sequences(df['summary'])

# Pad sequences
max_len_doc = 500  # Example max length for documents
max_len_summary = 50  # Example max length for summaries

document_padded = pad_sequences(document_sequences, maxlen=max_len_doc, padding='post')
summary_padded = pad_sequences(summary_sequences, maxlen=max_len_summary, padding='post')
```
